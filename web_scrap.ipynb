```python
from selenium import webdriver

from selenium.webdriver.chrome.options import Options

import csv
```

```python
#method for get user details usr_stars,usr_followers,usr_following
def get_user_details(sub_driver , usr_url):

    sub_driver.get(usr_url)

    main_div = sub_driver.find_element_by_class_name('UnderlineNav-body')

    sub_divs = main_div.find_elements_by_tag_name('a')

    usr_stars = '0'

    usr_followers = '0'

    usr_following = '0'

    try:
        usr_stars = str(sub_divs[3].find_element_by_class_name('Counter').text)

        print('Starts :'+ usr_stars)

    except Exception as exception:

        usr_stars = '0'

        print(exception)

    try:

        usr_followers = str(sub_divs[-2].find_element_by_class_name('Counter').text)

        print('Followers :' + usr_followers)

    except Exception as exception:

        usr_followers = '0'

        print(exception)

    try:
        usr_following = str(sub_divs[-1].find_element_by_class_name('Counter').text)

        print('Following :' + usr_following)

    except Exception as exception:

        usr_following = '0'

        print(exception)


    return usr_stars,usr_followers,usr_following

# set up option for web driver
options = Options()

#without opening chrome
options.add_argument('--headless')

options.add_argument('--no-sandbox')

#driver for get user details usr_stars,usr_followers,usr_following
#set chromedriver path
sub_driver = webdriver.Chrome(options=options, executable_path=r'\Users\ravichandrachilupuri\Downloads\chromedriver')

#driver for get main page details USER URL and USER NAME
#set chromedriver path
driver = webdriver.Chrome(options=options, executable_path=r'\Users\ravichandrachilupuri\Downloads\chromedriver')
#base URL
base_url = 'https://github.com/search?q=location%3Amelbourne&p=PAGENUMBER'
#go to first page
driver.get('https://github.com/search?q=location%3Amelbourne&p=1')
#get total count of user
usr_count = int(str(driver.find_element_by_class_name('codesearch-results').find_element_by_tag_name('h3').text).replace('users','').replace(',',''))
print('USER COUNT :'+str(usr_count))

total_page = int(usr_count / 10)

print('total Page :'+str(total_page))

user_details_list =[]
#loop for pagination
for index in range(2,3,1):
    print('-----------------------'+str(index)+'----------------------------------')
# replace page number for pagination
driver.get(base_url.replace('PAGENUMBER',str(index)))
main_div = driver.find_element_by_class_name('user-list')
sub_divs = main_div.find_elements_by_class_name('user-list-item')
print('Total USER :' + str(len(sub_divs)))
for sub_div in sub_divs:
    user_details = []
    usr_stars = usr_followers = usr_following = 'NONE'
    try:
        usr_url = str(sub_div.find_element_by_class_name('user-list-info').find_element_by_tag_name('a').get_attribute('href'))
        print('USER URL :' + usr_url)
        usr_stars, usr_followers, usr_following = get_user_details(sub_driver, usr_url)
    except Exception as exception:
        usr_url = 'None'
        print(exception)
    try:
        usr_name = str(sub_div.find_element_by_class_name('user-list-info').find_element_by_class_name('d-block').text)
        print('USER NAME :' + usr_name)
    except Exception as exception:
        usr_name = usr_url.split('/')[-1]
        print('USER NAME :' + usr_name)
        print(exception)
    try:
        usr_location = str(sub_div.find_element_by_tag_name('ul').find_elements_by_tag_name('li')[0].text)
        print('Location : ' + usr_location)
    except Exception as exception:
        usr_location = 'None'
        print(exception)
        user_details.append(usr_url)
        user_details.append(usr_name)
        user_details.append(usr_location)
        user_details.append(usr_stars)
        user_details.append(usr_followers)
        user_details.append(usr_following)
        user_details_list.append(user_details)
        print(user_details_list)
        # write into CSV File
        for use_detail in user_details_list:
            with open('user_details.csv', 'a') as csvFile:
                writer = csv.writer(csvFile)
                writer.writerow(use_detail)
                user_details_list = []
                #closing driver
driver.quit()

#closing driver
sub_driver.quit()
```
